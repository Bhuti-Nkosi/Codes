```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing a data from a text file(or csv file in general)
df = pd.read_csv('titanic.txt',sep=',',index_col=0)
```

```python
#habing a look at the data
df 
```

```python
#now we wanna find if there is some duplicates data and remove them
df_new = df.drop_duplicates()
df_new
```

```python
#Finding if we have blank entries. This is done to check whether the missing entries would be significant.
df.count()
```

```python
#Now I remove all blank entries
df_new = df.dropna()
df_new
```

```python
# Now that all the blanks are removed, we have a look at the statistics.
#This is done only on age because it is the only numerical data given.
df_new.Age.describe()
```

```python
#Finding total number of survivors
#This is the number of people who made it out alive of the titanic rackage.
df_new.Survived.sum()
```

```python
#Number of females and males
female = df_new.SexCode.sum()
male = 756 - female
print(male,female)
```

```python
#Extracted males who boarded the titanic and were from the 1st class
df_new2 = df_new.query('SexCode==0 and PClass =="1st"')
df_new2
```

```python
#All passengers who were over 60 and female
df_n = df_new.query('Age>60 and SexCode == 1')
df_n
```

```python
# A histogram to calculate number of people per each age group
df_new.Age.plot.hist('Age')
plt.xlabel('Age(years)')
plt.ylabel('Number of People')
plt.show()
```
